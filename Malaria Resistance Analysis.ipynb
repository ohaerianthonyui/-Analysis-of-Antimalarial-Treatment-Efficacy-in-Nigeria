```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\SD\Downloads\malaria_data_Nigeria.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
## Replacing spaces with _"
df.columns = df.columns.str.replace(' ', '_')
```

```python
##Verifying
df.columns
```

```python
## Changing datatypes
df['Sex'] = df['Sex'].astype('category')
df['State'] = df['State'].astype('category')
df['Age_Group'] = df['Age_Group'].astype('category')

# 3. CONVERT 'Treatment Start Date' and 'Treatment End Date' columns to datetime type
df['Treatment_Start_Date'] = pd.to_datetime(df['Treatment_Start_Date'])
df['Treatment_End_Date'] = pd.to_datetime(df['Treatment_End_Date'])
```

```python
df.dtypes
```

```python
df.head()
```

```python
##Checking for null values
df.isna().sum()
```

```python
## Filling Resistant type null values with "None"
df['Resistance_Type'].fillna('none', inplace=True)
```

```python
df.isna().sum()
```

```python
## Descriptive staistics for numerical columns
df.describe()
```

1. What is the average age of patients in the dataset?

```python

# Calculate the average age of patients
average_age = df['Age'].mean()
print(f"Average Age of Patients: {average_age}")
```

```python
# Plot the distribution of ages
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of Patients\' Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

2. What are the most commonly used drugs for treatment in the dataset?


```python
# Plot the most commonly used drugs
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Drug_Used', palette='Set2')
plt.title('Most Common Drugs Used for Treatment')
plt.xlabel('Drug Used')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

3. How many patients belong to each age group?


```python
# Plot the number of patients in each age group
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Age_Group', palette='Set1')
plt.title('Number of Patients by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

4. What is the average drug efficacy by region?


```python
# Plot the average drug efficacy by region
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Region', y='Drug_Efficacy_(%)', palette='coolwarm')
plt.title('Average Drug Efficacy by Region')
plt.xlabel('Region')
plt.ylabel('Average Drug Efficacy (%)')
plt.xticks(rotation=45)
plt.show()
```

5. What is the percentage of patients with drug resistance in each region?


```python
# Calculate the percentage of resistance by region
resistance_by_region = df[df['Resistance_Test_Result'] == 'Resistant'].groupby('Region').size() / df.groupby('Region').size() * 100

# Plot the resistance percentage by region
plt.figure(figsize=(12, 6))
resistance_by_region.plot(kind='bar', color='lightcoral')
plt.title('Percentage of Patients with Drug Resistance by Region')
plt.xlabel('Region')
plt.ylabel('Percentage of Resistance (%)')
plt.xticks(rotation=45)
plt.show()
```

6. What is the treatment outcome for each drug used?


```python
# Plot treatment outcomes by drug used
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Drug_Used', hue='Treatment_Outcome', palette='Set3')
plt.title('Treatment Outcome by Drug Used')
plt.xlabel('Drug Used')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

7. What are the most common symptoms experienced by patients?


```python
from wordcloud import WordCloud

# Combine all symptoms into one string
all_symptoms = df['Symptoms'].str.cat(sep=', ')

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_symptoms)

# Plot the word cloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Symptoms')
plt.show()
```

8. What is the average drug efficacy for male vs female patients?


```python
# Plot the average drug efficacy by sex
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Sex', y='Drug_Efficacy_(%)', palette='muted')
plt.title('Average Drug Efficacy by Sex')
plt.xlabel('Sex')
plt.ylabel('Average Drug Efficacy (%)')
plt.show()
```

9. What is the correlation between drug efficacy and age?


```python
# Calculate the correlation matrix
corr_matrix = df[['Age', 'Drug_Efficacy_(%)']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, linewidths=0.5)
plt.title('Correlation Between Age and Drug Efficacy (%)')
plt.show()
```

10. What is the treatment outcome by resistance type?


```python
# Plot treatment outcome by resistance type
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Resistance_Type', hue='Treatment_Outcome', palette='coolwarm')
plt.title('Treatment Outcome by Resistance Type')
plt.xlabel('Resistance Type')
plt.ylabel('Count')
plt.show()
```

11. How many patients had successful treatment based on the diagnosis method?


```python
# Plot successful treatments by diagnosis method
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['Treatment_Outcome'] == 'Cured'], x='Diagnosis_Method', palette='Set2')
plt.title('Successful Treatments by Diagnosis Method')
plt.xlabel('Diagnosis Method')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

12. What is the average treatment duration for patients (in days)?


```python
df['Treatment_Duration'] = (df['Treatment_End_Date'] - df['Treatment_Start_Date']).dt.days
df = df[df['Treatment_Duration'] >= 0]
mean_duration = df['Treatment_Duration'].mean()
```

```python
print(f"Mean Treatment Duration: {mean_duration} days")
```

```python
# Plot the distribution of treatment duration
plt.figure(figsize=(10, 6))
sns.histplot(df['Treatment_Duration'], kde=True, bins=20, color='salmon')
plt.title('Distribution of Treatment Duration (Days)')
plt.xlabel('Treatment Duration (Days)')
plt.ylabel('Frequency')
plt.show()
```

13. What is the treatment outcome by state?


```python
# Plot the treatment outcome distribution by state
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='State', hue='Treatment_Outcome', palette='Set1')
plt.title('Treatment Outcome by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

14. What is the distribution of drug efficacy across different resistance test results?


```python
# Box plot to compare drug efficacy by resistance test result
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Resistance_Test_Result', y='Drug_Efficacy_(%)', palette='coolwarm')
plt.title('Drug Efficacy by Resistance Test Result')
plt.xlabel('Resistance Test Result')
plt.ylabel('Drug Efficacy (%)')
plt.show()
```

15. Which symptoms are most commonly associated with resistance to treatment?


```python
# Extract symptoms for resistant patients
resistant_patients = df[df['Resistance_Test_Result'] == 'Resistant']
resistant_symptoms = resistant_patients['Symptoms'].str.cat(sep=', ')

# Generate word cloud
wordcloud_resistant = WordCloud(width=800, height=400, background_color='white').generate(resistant_symptoms)

# Plot the word cloud for resistant patients
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud_resistant, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Symptoms in Resistant Patients')
plt.show()
```
